## BeautifulSoup

```python
response = requests.get('https://sports.daum.net/soccer')
response.status_code
```

```python
html = response.text
soup = BeautifulSoup(html, 'lxml')
```

```python
ol = soup.find_all('ol')[0]
```

```python
lis = ol.find_all('li')
popular_news =[]

for li in lis:
    rank = li.find('em').text
    title = li.find('a').text
    link = li.find('a')['href']
    
    popular_news.append({'rank' : rank, 'title' : title, 'link' : link})

popular_news
```

```python
##최종정리##

response = requests.get('https://sports.daum.net/soccer') # requests로 분석할 문서 텍스트 요청
html = response.text # 응답 텍스트를 변수에 할당

soup = BeautifulSoup(html, 'lxml') # beautifulsoup 을 이용한 html 분석 시작(lxml 이용)
ol = soup.find_all('ol')[0] # 리스트 중 0번째 ol 태그 선택(많이 본 뉴스)
lis = ol.find_all('li') # ol태그의 뉴스단위는 li로 구분되므로 모든 li를 찾아 구분한 뉴스 묶음 lis
popular_news = []
# 뉴스 한단위(li) 하나하나에 대해
for li in lis:
    rank = li.find('em').text
    title = li.find('a').text
    news_link = li.find('a')['href']
    popular_news.append({
        'rank': rank,
        'news_title': title,
        'news_link': news_link,
    }) # dict 형태로 한단위의 뉴스를 popular_news라는 리스트에 추가
popular_news
```

```python
## 최종정리 ##
response = requests.get('https://www.rottentomatoes.com/')
response.status_code

html = response.text

soup = BeautifulSoup(html, 'lxml')
box_office = soup.find_all('section', attrs={'class':"dynamic-text-list"})[-1]
box_office.find('h2')

chart_name = box_office.find('h2').text.strip('\n')

lis = box_office.find_all('li')
result = []
for li in lis:
    title = li.select_one('a:nth-child(1)').text.strip('\n')
    score = li.select_one('a:nth-child(2)').text.strip('\n ')
    result.append({
        'title': title,
        'score': score if len(score) > 0 else '-',
    })
result
```

## Practice

영화제목, 개봉일, 이미지링크 가져오기

```python
response = requests.get('https://www.rottentomatoes.com/browse/movies_in_theaters/sort:popular')
response.status_code
html = response.text
soup = BeautifulSoup(html,'lxml')

movie = []
url = 'https://www.rottentomatoes.com'

divs = soup.find_all('div', attrs ={'class':'js-tile-link'})

for div in divs:
    img = div.find('rt-img')['src']
    title = div.find('span', attrs={'class':"p--small"}).text.strip('\n ')
    open_date = div.find('span', attrs={'class':"smaller"}).text.strip('\n ')
    link = url + div.find('a')['href']
    movie.append({'title':title, 'open_date':open_date, 'link':link, 'img':img})

movie
```

```python
response = requests.get('https://www.rottentomatoes.com/browse/movies_in_theaters/sort:popular')
response.status_code
html = response.text
soup = BeautifulSoup(html,'lxml')

movie = []
url = 'https://www.rottentomatoes.com'

divs = soup.find_all('div', attrs ={'class':'js-tile-link'})

for div in divs:
    img = div.find('rt-img')['src']
    title = div.select('span.p--small')[0].string.strip("\n ")
    open_date = div.find('span', attrs={'class':"smaller"}).text.strip('\n ')
    link = url + div.find('a')['href']
    movie.append({'title':title, 'open_date':open_date, 'link':link, 'img':img})

movie
```
